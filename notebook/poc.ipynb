```python
import importlib
import pkgutil


def list_modules(package_name):
    package = importlib.import_module(package_name)
    if package is None:
        print("Package not found.")
        return

    print(f"Modules in package '{package_name}':")
    package_path = package.__path__

    for _, module_name, _ in pkgutil.iter_modules(package_path):
        print(module_name)


# Replace 'your_package' with the package name you are interested in
list_modules("ai_data_science_team")
print("\n================\n")
list_modules("pydantic_ai")
```

```python
import inspect


def list_functions(module_name):
    module = importlib.import_module(module_name)
    functions = inspect.getmembers(module, inspect.isfunction)

    print(f"Functions in module '{module_name}':")
    for function_name, _ in functions:
        print(function_name)


# Replace 'your_module' with the module name you are interested in
list_functions("ai_data_science_team")
```
